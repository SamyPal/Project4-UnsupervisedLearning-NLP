```python
data = json.load(open('blogs_0000001.json'))
df = json_normalize(data)
df = pd.DataFrame(df)
df.head()
```

```python
def file_import(number_articles): 
    df=[]
    for i in range(2, number_articles+2):
        j = str(i).zfill(7)
        try:
            data = json.load(open(base_path+'news_'+j+'.json'))
            norm_data = json_normalize(data)
            df.append(pd.DataFrame(norm_data))
        except Exception as e:
            print(e)  
    df_news = pd.concat(df)
    print(df_news.shape)
    return df_news
```

```python
number_articles = 57455
df = file_import(number_articles)
df = df.reset_index()
df.columns
```

```python
cols = ['title', 'author', 'text', 'thread.spam_score', 'published']
df = df[cols]
df.head()
```

```python
df[['Date', 'Time']] = df.published.str.split(pat='T', expand=True)
df['Time'] = df.Time.str.split(pat='+', n=0, expand=True)
df['Time'] = df.Time.str.split(pat='.', n=0, expand=True)
```

```python
df.head()
```

```python
for i in range(len(df)-1):
    df.loc[i,'Date'] = dt.datetime.strptime(df.loc[i,'Date'],'%Y-%m-%d').date()
    df.loc[i,'Time'] = dt.datetime.strptime(df.loc[i,'Time'], '%H:%M:%S').time()
```

```python
df.head()
```

```python
def add_senti(df):
    senti = []
    for row in range(len(df)-1):
        senti.append(sia.polarity_scores(df.loc[row,'text']))
    senti_df = pd.DataFrame(senti)
    result_df = pd.concat([df, senti_df], join='outer', axis=1)
    return result_df
```

```python
df_senti = add_senti(df)
df_senti.head()
```

```python
df_summary = df_senti.groupby(['Date'])['compound'].mean().reset_index()
```

```python
df_summary.head(20)
```

```python
plt.rc('lines', linewidth=4)
plt.figure(figsize=(12,6))

x=df_summary.loc[4:,'Date']
y=df_summary.loc[4:,'compound'] 
plt.plot(x, y, label='Overall_Fin_Sentiment')
plt.legend(fontsize='16')
plt.ylabel('Sentiment', fontsize='16')
plt.xlabel('Date', fontsize='16')
plt.title('Sentiment Analysis for Reddit Financial Data')
plt.show()
```
